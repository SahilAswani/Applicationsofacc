```python
class Particle:
    def __init__(self, Z, A):
        self.Z = Z
        self.A = A
    def number_of_neutrons(self):
        return self.A - self.Z
    
#Example:Iodine-131
iodine_131 = Particle(Z=53, A=131)
neutrons_in_iodine_131 = iodine_131.number_of_neutrons()
neutrons_in_iodine_131
    
```
